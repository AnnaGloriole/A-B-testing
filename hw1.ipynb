### Задание 2. Сформулируйте 5-10 гипотез для выбранного продукта.

*	##### Создание уникального упрощенного эл. каталога для конкретного пользователя
 в зависимости от покупок / предпочтений / с наиболее часто покупаемыми товарами, чтобы все было “под рукой”. Для постоянных покупателей только с опред. товарами. Для сравнения приводить товары 2х – 3х брендов, которые ранее не покупались. Или 2-3х сопутствующих товаров из категории. 
 Для того, чтобы страница предложений была не перегружена и выбор было сделать проще.
*	##### Смс–уведомление о снижении цены товара в корзине 
Для возможности оперативно сделать покупку по низкой цене. На товары от определенной суммы при подписке пользователя на данное смс. 
*	##### Возможность переслать “корзину” на другой аккаунт 
*   ##### "Совместная" корзина 
для упрощения совместных покупок с разных аккаунтов 
*	##### При заказе с сайта продукта более 3 раз 
Предоставление скидки на этот конкретный продукт
*	##### Промо новинки-подарки для постоянных покупателей опред. категорий/брендов.
В качестве тестовых образцов при общей сумме заказа на какую-либо сумму.
*	##### Устанавливай скидку на свою категорию. 
Для самостоятельной настройки пользователем (предоставление скидки на категорию 1 период в год/ опред. кол-во покупок в год). Скидка варьируется в зависимости от категории, периода и кол-ва покупок. Аналог - выбор кредита в банке.
